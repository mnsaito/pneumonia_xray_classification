```python
# Standard Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
import os
import glob
from PIL import Image, ImageStat

# Modeling Imports
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import Sequential
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dense, Flatten, Input, Conv2D, MaxPooling2D

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, recall_score
from sklearn.utils import class_weight

# Lime Imports
%load_ext autoreload
%autoreload 2
import lime
from lime import lime_image
from skimage.segmentation import mark_boundaries

import sys
sys.path.append('../../src')
import functions as fx
```

```python
train_path = '../../chest_xray/train'
test_path = '../../chest_xray/new_test'
val_path = '../../chest_xray/new_val'
# Include data augmentation with training data, and 
train_generator = ImageDataGenerator(rescale = 1./255.,
                                    rotation_range = 30, zoom_range = [0.9, 1.2],
                                    width_shift_range = 0.1, height_shift_range = 0.1,
                                    horizontal_flip = True) \
    .flow_from_directory(train_path, target_size = (256, 256), 
                         batch_size = 64, class_mode = 'binary')
test_generator = ImageDataGenerator(rescale = 1./255.) \
    .flow_from_directory(test_path, target_size = (256, 256), 
                         batch_size = 64, class_mode = 'binary')
val_generator = ImageDataGenerator(rescale = 1./255.) \
    .flow_from_directory(val_path, target_size = (256, 256), 
                         batch_size = 64, class_mode = 'binary')
```

```python
# Calculate weights to balance normal and pneumonia data in training set
class_weights = class_weight.compute_class_weight(
    class_weight = 'balanced', classes = np.unique(train_generator.classes), 
    y = train_generator.classes)
train_class_weights = dict(enumerate(class_weights))
```

```python
# Stop training if val_loss does not decrease at least 0.01 in 5 epochs
early_stop_loss = EarlyStopping(monitor = 'val_loss', 
                                min_delta = 0.01, patience = 5, 
                                mode = 'min', restore_best_weights = True)
# Some of the models predicted all images were pneumonia
# and consistently resulted in val_accuracy=0.6034 and val_recall=1
# To break this loop, stop training if val_accuracy does not increase at least 0.01 in 5 epochs
early_stop_acc = EarlyStopping(monitor = 'val_accuracy', 
                               min_delta = 0.01, patience = 5, 
                               mode = 'max', restore_best_weights = True)
```

### Baseline Model with No Convolutional Layers

```python
def baseline_model(image):
    layers = [
        Input(image.shape),
        Flatten(),
        Dense(128, activation = 'relu'),
        Dense(1, activation = 'sigmoid')
    ]
    model = Sequential(layers)
    model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy', 'Recall'])
    return model
```

```python
base_model = baseline_model(train_generator[0][0][0])
base_model.summary()
```

```python
base_result = base_model.fit(train_generator, epochs = 10, validation_data=val_generator,
                             class_weight = train_class_weights, 
                             callbacks = [early_stop_loss, early_stop_acc])
```

### Results
- As depicted below, model is over-fitting on training data
- Model is under-predicting pneumonia cases
- Model is not focusing on lungs to classify pneumonia

```python
fx.visualize_training_results(base_result)
```

```python
fx.create_confusion_matrix(base_model, val_generator)
```

```python
fx.display_lime(base_model, val_generator)
```

### Baseline Model with Single Convolutional Layer

```python
def baseline_cnn(image):
    layers = [
        Input(image.shape),
        Conv2D(32, (3,3), activation = 'relu'),
        Flatten(),
        Dense(128, activation = 'relu'),
        Dense(1, activation = 'sigmoid')
    ]
    model = Sequential(layers)
    model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy', 'Recall'])
    return model
```

```python
base_cnn = baseline_cnn(train_generator[0][0][0])
base_cnn.summary()
```

```python
cnn_result = base_cnn.fit(train_generator, epochs = 10, validation_data=val_generator,
                        class_weight = train_class_weights, 
                        callbacks = [early_stop_loss, early_stop_acc])
```

### Results
- As depicted below, addiing a convolutional layer significantly improved model performance
- Model is not focusing on lungs to classify pneumonia

```python
fx.visualize_training_results(cnn_result)
```

```python
fx.create_confusion_matrix(base_cnn, val_generator)
```

```python
fx.display_lime(base_cnn, val_generator)
```
